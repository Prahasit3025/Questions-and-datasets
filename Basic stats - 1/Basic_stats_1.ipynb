#  Descriptive Analytics for Numerical Columns

```python
df = pd.read_csv(list(uploaded.keys())[0])
print(df.head())
numerical_columns = df.select_dtypes(include=[np.number]).columns.tolist()
print("Numerical Columns:", numerical_columns)
statistics = df[numerical_columns].describe().T
print("Basic Statistics for Numerical Columns:\n", statistics)

# Calculate mean, median, mode, and standard deviation
for col in numerical_columns:
    print(f"--- {col} ---")
    print(f"Mean: {df[col].mean()}")
    print(f"Median: {df[col].median()}")
    print(f"Mode: {df[col].mode()[0]}")
    print(f"Standard Deviation: {df[col].std()}\n")
```

#Data Visualization

```python
print("Histogram:")
for col in numerical_columns:
    plt.figure(figsize=(8,4))
    plt.hist(df[col], bins=20, color='c', edgecolor='black')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
print("Boxplots:")
for col in numerical_columns:
    plt.figure(figsize=(8,4))
    sns.boxplot(df[col], color='skyblue')
    plt.title(f'Boxplot for {col}')
    plt.grid(True)
    plt.show()
```

# Bar Chart for Categorical Columns

```python
categorical_columns = df.select_dtypes(include=[object]).columns.tolist()
print("Categorical Columns:", categorical_columns)
for col in categorical_columns:
    plt.figure(figsize=(8,4))
    sns.countplot(data=df, x=col, palette='Set2')
    plt.title(f'Count of Categories in {col}')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()
```

# Standardization of Numerical Variables

```python
scaler = StandardScaler()
df_scaled = df.copy()
df_scaled[numerical_columns] = scaler.fit_transform(df[numerical_columns])
print("Before Standardization (First 5 rows):\n", df[numerical_columns].head())
print("\nAfter Standardization (First 5 rows):\n", df_scaled[numerical_columns].head())
for col in numerical_columns:
    plt.figure(figsize=(8,4))
    plt.hist(df_scaled[col], bins=20, color='g', edgecolor='black')
    plt.title(f'Distribution of {col} after Standardization')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

# Convert Categorical Data into Dummy Variables

```python
df_encoded = pd.get_dummies(df_scaled, columns=categorical_columns)
print("Transformed Dataset (First 5 rows):\n", df_encoded.head())
```
